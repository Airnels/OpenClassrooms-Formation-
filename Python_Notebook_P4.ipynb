### <font color=#009F0C> B - Nombre théorique de personnes qui pourraient être nourries

&nbsp;
    
**Pour répondre à cette demande, je ne dispose que des données de disponibilité alimentaire pour l'année 2013 ce qui m'empêchera d'avoir un résultat fiable pour l'année 2017**
    
**Je commence dans un premier temps par grouper par pays "dispo". Cela me permet d'avoir la somme de "Disponibilité alimentaire (Kcal/personne/jour)" par pays**      

```python
dispo.head()
```

```python
dispo_pays = dispo.groupby("Pays").sum()
dispo_pays.head()
```

**Pour obtenir la population potentiellement nourrie, je vais suivre les étapes suivantes :**

- Ayant la disponibilité alimentaire par personne, je fixe un indicateur moyen de 2250 Kcal/jour pour considérer qu'une personne est nourrie (moyenne de 2000/femme et 2500/homme)


- Avec cet indicateur, je vais pouvoir calcuer le taux de personnes nourries par pays


- Ce taux sera ensuite multiplié par la population pour me donner une population potentiellement nourrie par pays


- Puis la somme de ces populations potentiellement nourries me donnera une population globale

&nbsp;

**Le résultat me donnera le nombre pour l'année 2013. Pour le nombre théorique pour l'année 2017, il y aura 2 solutions :**

- **Cas 1 :** on considère que la disponibilité alimentaire totale par pays (Kcal/jour) est inchangée


- **Cas 2 :** on considère que la disponibilité alimentaire totale par pays (Kcal/jour) a évolué au même rythme que la population (c'est-à-dire que la disponibilité alimentaire en Kcal/pers/jour est inchangée)

&nbsp;

**Avant de commencer les manipulations sur les dataframes, je vais d'abord m'attarder sur ces 2 cas** 



&nbsp;

DA_tot (disponibilité alimentaire totale en kcal/jour) = DA_pers (disponibilité alimentaire en kcal/pers/jour) * Pop (Population)

Taux_pop_nour (Taux de population nourrie) = DA_pers / 2250

Pop_pot_nour (Population potentiellement nourrie) = Pop * Taux_pop_nour

&nbsp;

**Cas 1 : La disponibilité alimentaire totale par pays est inchangée**

&nbsp;

DA_tot_2013 = DA_tot_2017 = DA_pers_2013 * Pop_2013 = DA_pers_2017 * Pop_2017 

Taux_pop_nour_2013 = DA_pers_2013 / 2250 

Taux_pop_nour_2017 = DA_pers_2017 / 2250

Pop_pot_nour_2013 = Pop_2013 * Taux_pop_nour_2013

Pop_pot_nour_2017 = Pop_2017 * Taux_pop_nour_2017 

                  = Pop_2017 * (Da_pers_2017 / 2250) 
                  
                  = Pop_2017 * (DA_pers_2013 * Pop_2013 / Pop_2017) / 2250
                  
                  = Taux_pop_nour_2013 * Pop_2013
                  
                  = Pop_pot_nour_2013


**Pour le cas 1, si la disponibilité alimentaire totale est inchangée, alors la population potentiellement nourrie sera également inchangée**

&nbsp;

**Cas 2 : La disponibilité alimentaire par personne est inchangée**

&nbsp;

DA_pers_2013 = DA_pers_2017

Taux_pop_nour_2013 = DA_pers_2013 / 2250 = DA_pers_2017 / 2250 = Taux_pop_nour_2017 

Pop_pot_nour_2013 = Pop_2013 * Taux_pop_nour_2013

Pop_pot_nour_2017 = Pop_2017 * Taux_pop_nour_2017

                  = Pop_2017 / Pop_2013 * Pop_pot_nour_2013
                  

**Pour le cas 2, si la disponibilité alimentaire par personne est inchangée, alors la population potentiellement nourrie en 2017 sera celle de 2013 multipliée par (Popuplation de 2017 / Population de 2013)** 
                  
&nbsp;

**Maintenant, j'ai une vision plus claire des données qui vont m'être utiles pour répondre à la demande de Marc**

**Je vais devoir faire une jointure de "dispo_pays" avec "pop" pour les années 2013 et 2017**

**Je vais maintenant ajouter les colonnes pour le taux de population nourrie et la population potentiellement nourrie en 2013 ( = 2017 pour le cas 1) et en 2017 pour le cas 2**
