## <font color=#2921CA> II - Jointure entre les 3 dataframes
    
&nbsp; 
    

Les premières modifications étant faites sur mes dataframes, je vais maintenant procéder à la jointure de ceux-ci afin de créer un unique dataframe exploitable pour mes analyses 

Précèdemment, on avait pu voir que "*transa*" possèdait la variable **client_id** en commun avec "*custo*"  et la variable **id_prod** en commun avec "*prod*"

**Client_id** est unique dans "*custo*", idem pour **id_prod** dabs "*prod*", je vais donc pouvoir joindre ces 2 dataframes à "*transa*"

Je n'ai besoin que des données relatives à la boutique en ligne donc les 2 jointures se feront du coté de "*transa*"

```python
tc = pd.merge(t1, c1, on = 'client_id', how = 'left')
tcp = pd.merge(tc, p1, on ='id_prod', how ='left')
```

Seul le produit '0_2245' présente des valeurs nulles pour **price** et **categ**. Je vais devoir trouver une solution pour remplacer ces valeurs nulles

Pour **categ** : j'ai remarqué que les valeurs de **id_prod** sont de la forme : "nombre1_nombre2" et que nombre1 semble correspondre à la valeur de **categ**

On va vérifier si cela est le cas

En effet nombre1 correspond toujours à la catégorie, donc le produit '0_2245' a pour catégorie : 0

Maintenant je dois trouver une solution pour estimer le prix de ce produit en limitant le risque d'erreurs que cela pourrait entrainer pour la partie analyse

Comme je n'ai pas la possibilité de trouver son prix réél, une imputation par la moyenne peut-être une solution

```python
p1.describe()
```

```python
p1.loc[p1['categ'] == 0].describe()
```

```python
p1.loc[p1['categ'] == 1].describe()
```

```python
p1.loc[p1['categ'] == 2].describe()
```

En regardant les indicateurs numériques du prix en général et par catégorie, je remarque que ces indicateurs ont des valeurs grandement différentes selon la catégorie, en effet les indicateurs du prix de la catégorie 0 sont inférieurs à ceux de la catégorie 1 qui sont inférieurs à ceux de la catégorie 2

Je décide alors de remplacer le prix du produit '0_2245' par celui de la moyenne des produits de catégorie 0

```python
tcp['categ'].loc[tcp['id_prod'] == '0_2245'] = 0
moy0 = p1['price'].loc[p1['categ'] == 0].mean()
tcp['price'].loc[tcp['id_prod'] == '0_2245'] = round(moy0,2)
```

## <font color=#2921CA> III - Ajout de variables supplémentaires
    
&nbsp;    

Avant de basculer dans le notebook de la partie 2, j'ai décidé d'ajouter à mon dataframe des variables qui me seront utiles pour la partie suivante :

- **age** (calculant l'age au 31/12/2022 à partir de l'année de naissance)

- **tranche_age** (regroupant les ages dans différentes classes d'ages)

- **mois** (regroupant les dates des transactions par mois)

- **jour** (regroupant les dates des transactions par jour)

- **jours_achat** (caclulant le nombre de jours entre la transaction la plus ancienne et la transaction la plus récente d'un client)

- **statut** (indiquant le statut d'un client : professionnel ou particulier)

- **nombre** (égale à 1 pour toutes les lignes, elle servira à calculé rapidement le nombre d'itération lors des groupby sommés)

Ces variables proviennent à la base de mes recherches de la partie analyse, afin d'épurer celui-ci j'ai décidé d'ajouter ces variables à la fin de ce notebook

```python
tcp['age'] = 2022 - tcp['birth']
tranche_age = ['18-30','31-50','51-70','71-99']
tcp['tranche_age'] = pd.cut(tcp['age'],bins=[18,31,51,71,100], right=False,labels=tranche_age)
tcp['mois'] = tcp['date'].dt.to_period('M')
tcp['jour'] = tcp['date'].dt.to_period('D')
tcp['jours_achat'] = 0
for i in tcp['client_id'].unique(): 
    tcp['jours_achat'].loc[tcp['client_id']==i] = max(tcp['jour'].loc[tcp['client_id']==i])-min(tcp['jour'].loc[tcp['client_id']==i])
tcp['jours_achat'] = tcp['jours_achat']/np.timedelta64(1,'D')
pro = ['c_1609','c_4958','c_6714','c_3454']
tcp['statut'] = ["professionnel" if i in pro else "particulier" for i in tcp['client_id']]
tcp['nombre'] = 1
tcp = tcp[['session_id','date','mois','jour','id_prod','price','categ','client_id','sex','birth','age','tranche_age','jours_achat','statut','nombre']]
tcp.head()
```
